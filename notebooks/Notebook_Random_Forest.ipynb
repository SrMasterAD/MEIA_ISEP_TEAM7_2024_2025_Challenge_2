```python
# File: wine_classification_optimized.py

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import logging
import joblib
import shap
from sklearn.model_selection import train_test_split, RandomizedSearchCV, GridSearchCV, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.metrics import classification_report, accuracy_score, f1_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.utils import shuffle
from tabulate import tabulate
from sklearn.model_selection import learning_curve

# Setup Logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')


def log_table(df: pd.DataFrame, message: str, limit=5):
    """
    Log a preview of a DataFrame in a table format.
    """
    logging.info(f"{message}\n{tabulate(df.head(limit), headers='keys', tablefmt='pretty')}")


def load_data(filepath: str) -> pd.DataFrame:
    """
    Load dataset from the specified file path.
    """
    try:
        df = pd.read_csv(filepath)
        log_table(df, f"Dataset loaded: {df.shape[0]} rows, {df.shape[1]} columns.")
        return df
    except Exception as e:
        logging.error(f"Error loading dataset: {e}")
        raise


def explore_data(df: pd.DataFrame):
    """
    Perform data exploration with detailed logs and visualizations.
    """
    logging.info("Exploring the dataset...")
    logging.info(f"Missing values:\n{df.isnull().sum()}")
    log_table(df.describe(), "Summary statistics:")

    # Convert unique value counts to a DataFrame for logging
    unique_values = df.nunique().reset_index()
    unique_values.columns = ['Column', 'Unique Values']
    log_table(unique_values, "Unique values per column:")

    # Class distribution
    plt.figure(figsize=(10, 5))
    sns.countplot(data=df, x='Type', order=df['Type'].value_counts().index)
    plt.title("Distribution of Wine Types")
    plt.ylabel("Count")
    plt.xlabel("Wine Type")
    plt.show()


def preprocess_data(df: pd.DataFrame) -> pd.DataFrame:
    """
    Preprocess the dataset by handling missing values, dimensionality reduction, and encoding.
    Args:
        df (pd.DataFrame): Input dataset.
    Returns:
        pd.DataFrame: Preprocessed dataset.
    """
    # Handle missing values
    logging.info("Handling missing values...")
    df.fillna({
        'Type': df['Type'].mode()[0],  # Fill missing 'Type' with mode
        'Grapes': "['Unknown']"  # Fill missing 'Grapes' with 'Unknown'
    }, inplace=True)

    # Drop irrelevant columns
    irrelevant_columns = [
        'WineName', 'ABV', 'WineID', 'WineryName', 'Website', 'Vintages', 'Elaborate', 'Body', 'Acidity', 'Code', 'Country', 'RegionID', 'RegionName', 'WineryID'
    ]
    df.drop(columns=irrelevant_columns, inplace=True)

    # Process 'Grapes' column
    logging.info("Processing 'Grapes' column...")
    df['Grapes'] = df['Grapes'].apply(lambda x: eval(x) if isinstance(x, str) else x)
    grape_counts = pd.Series([grape for grapes in df['Grapes'] for grape in grapes]).value_counts()
    top_grapes = grape_counts.head(10).index
    df['Processed_Grapes'] = df['Grapes'].apply(lambda x: [grape if grape in top_grapes else 'Other' for grape in x])

    for grape in top_grapes:
        df[f'Grape_{grape}'] = df['Processed_Grapes'].apply(lambda x: 1 if grape in x else 0)
    df.drop(columns=['Grapes', 'Processed_Grapes'], inplace=True)

    # One-hot encode 'Type' column
    logging.info("One-hot encoding 'Type' column...")
    df = pd.get_dummies(df, columns=['Type'], drop_first=True)

    # Harmonize Dish Categories
    logging.info("Categorizing 'Harmonize' column...")
    dish_to_category = {
        'Beef': 'Meat', 'Lamb': 'Meat', 'Pork': 'Meat', 'Veal': 'Meat', 'Game Meat': 'Meat',
        'Duck': 'Meat', 'Ham': 'Meat', 'Cold Cuts': 'Meat', 'Cured Meat': 'Meat', 'Poultry': 'Poultry',
        'Chicken': 'Poultry', 'Rich Fish': 'Fish & Seafood', 'Lean Fish': 'Fish & Seafood',
        'Shellfish': 'Fish & Seafood', 'Seafood': 'Fish & Seafood', 'Soft Cheese': 'Cheese',
        'Hard Cheese': 'Cheese', 'Blue Cheese': 'Cheese', 'Goat Cheese': 'Cheese', 'Cheese': 'Cheese',
        'Pasta': 'Pasta', 'Pizza': 'Pasta', 'Vegetarian': 'Vegetarian & Vegan', 'Mushrooms': 'Vegetarian & Vegan',
        'Salad': 'Vegetarian & Vegan', 'Fruit': 'Vegetarian & Vegan', 'Dessert': 'Desserts', 'Cake': 'Desserts'
    }
    df['Harmonize'] = df['Harmonize'].apply(lambda x: [dish_to_category.get(dish, 'Other') for dish in eval(x)])
    dish_categories = list(set([cat for sublist in df['Harmonize'] for cat in sublist]))

    for category in dish_categories:
        df[f'Harmonize_{category}'] = df['Harmonize'].apply(lambda x: 1 if category in x else 0)
    df.drop(columns=['Harmonize'], inplace=True)

    # Correlation heatmap
    numeric_df = df.select_dtypes(include=['number'])  # Filter only numeric columns
    if numeric_df.shape[1] > 1:  # Check if there are enough numeric columns for correlation
        plt.figure(figsize=(12, 8))
        sns.heatmap(numeric_df.corr(), annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
        plt.title("Correlation Heatmap")
        plt.show()
    else:
        logging.warning("Not enough numerical columns for a correlation heatmap.")
    
    log_table(df, "Preview after preprocessing:")
    return df


def split_features_targets(df: pd.DataFrame):
    """
    Split dataset into features and targets.
    """
    X = df.filter(like='Harmonize_')
    y_type = df[[col for col in df.columns if col.startswith('Type_')]]
    y_grape = df[[col for col in df.columns if col.startswith('Grape_')]]
    return X, y_type, y_grape


def train_random_forest(X_train, y_train):
    """
    Train Random Forest with RandomizedSearchCV.
    """
    param_dist = {
        'estimator__n_estimators': [100, 200, 300],
        'estimator__max_depth': [None, 10, 20],
        'estimator__min_samples_split': [2, 5],
        'estimator__min_samples_leaf': [1, 2],
    }
    model = RandomizedSearchCV(
        estimator=MultiOutputClassifier(RandomForestClassifier(random_state=42)),
        param_distributions=param_dist, n_iter=10, cv=3, scoring='accuracy', random_state=42, n_jobs=-1
    )
    model.fit(X_train, y_train)
    logging.info(f"Best Parameters from RandomizedSearchCV: {model.best_params_}")
    return model.best_estimator_


def evaluate_model(model, X_test, y_test, label: str):
    """
    Evaluate the model with classification metrics and confusion matrix.
    """
    y_pred = model.predict(X_test)

    # Handle multi-output/multilabel targets
    if y_test.shape[1] > 1:  # Multi-output task
        for i, col in enumerate(y_test.columns):
            logging.info(f"Evaluating target: {col}")
            col_y_test = y_test.iloc[:, i]
            col_y_pred = pd.DataFrame(y_pred).iloc[:, i]
            logging.info(f"{col} Accuracy: {accuracy_score(col_y_test, col_y_pred)}")
            logging.info(f"{col} F1 Score: {f1_score(col_y_test, col_y_pred, average='weighted')}")
            logging.info(f"{col} Classification Report:\n{classification_report(col_y_test, col_y_pred)}")
            plot_confusion_matrix(col_y_test, col_y_pred, title=f"{label} - {col}")
    else:  # Single-label task
        y_test_labels = y_test.idxmax(axis=1)
        y_pred_labels = pd.DataFrame(y_pred).idxmax(axis=1)
        logging.info(f"{label} Accuracy: {accuracy_score(y_test_labels, y_pred_labels)}")
        logging.info(f"{label} F1 Score: {f1_score(y_test_labels, y_pred_labels, average='weighted')}")
        logging.info(f"{label} Classification Report:\n{classification_report(y_test_labels, y_pred_labels)}")
        plot_confusion_matrix(y_test_labels, y_pred_labels, title=f"{label} Confusion Matrix")


def plot_confusion_matrix(y_true, y_pred, title="Confusion Matrix"):
    """
    Plot confusion matrix with enhancements.
    """
    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(cmap="Blues", colorbar=True)
    plt.title(title)
    plt.show()


def plot_shap(model, X_train, target_names):
    """
    Generate SHAP plots for each target in a MultiOutputClassifier.
    Args:
        model: Trained MultiOutputClassifier.
        X_train: Training features.
        target_names: List of target names.
    """
    if not hasattr(model, "estimators_"):
        raise ValueError("The model must have `estimators_` attribute (e.g., MultiOutputClassifier).")
    
    for i, estimator in enumerate(model.estimators_):
        logging.info(f"Generating SHAP plots for target: {target_names[i]}")
        explainer = shap.TreeExplainer(estimator)
        shap_values = explainer.shap_values(X_train)

        # Summary plot (feature importance)
        shap.summary_plot(shap_values, X_train, plot_type="bar", show=False)
        plt.title(f"SHAP Summary (Bar) - {target_names[i]}")
        plt.show()

        # Detailed summary plot
        shap.summary_plot(shap_values, X_train, show=False)
        plt.title(f"SHAP Summary - {target_names[i]}")
        plt.show()


def plot_learning_curve(estimator, X, y, title):
    """
    Plot learning curves for the estimator.
    """
    train_sizes, train_scores, test_scores = learning_curve(estimator, X, y, cv=5, scoring='accuracy', n_jobs=-1)
    train_mean = train_scores.mean(axis=1)
    test_mean = test_scores.mean(axis=1)

    plt.figure(figsize=(10, 6))
    plt.plot(train_sizes, train_mean, label="Training Score", color="blue")
    plt.plot(train_sizes, test_mean, label="Validation Score", color="orange")
    plt.title(f"Learning Curve: {title}")
    plt.xlabel("Training Set Size")
    plt.ylabel("Accuracy Score")
    plt.legend(loc="best")
    plt.show()


if __name__ == "__main__":
    df = load_data("../datasets/XWines_Full_100K_wines.csv")
    explore_data(df)
    df = preprocess_data(df)
    X, y_type, y_grape = split_features_targets(df)

    # Split dataset
    X_train, X_test, y_train_type, y_test_type = train_test_split(X, y_type, test_size=0.3, random_state=42)
    _, _, y_train_grape, y_test_grape = train_test_split(X, y_grape, test_size=0.3, random_state=42)

    # Train models
    best_model_type = train_random_forest(X_train, y_train_type)
    best_model_grape = train_random_forest(X_train, y_train_grape)

    # Evaluate models
    evaluate_model(best_model_type, X_test, y_test_type, label="Type")
    evaluate_model(best_model_grape, X_test, y_test_grape, label="Grape")

    # SHAP and learning curve plots
    plot_shap(best_model_type, X_train, target_names=y_train_type.columns)
    plot_shap(best_model_grape, X_train, target_names=y_train_grape.columns)
```
